```python
%pylab inline
import sys
import xlrd
import pandas as pd
import numpy as np
import os
from os.path import expanduser
import matplotlib.pyplot as plt
import matplotlib.style
import math
import csv
```

```python
pwd = "Zid-Lab/Analysis/Polarity/Kyle/"
list_of_genes = {}
directory_list = [os.path.join(os.environ["HOME"],pwd+"Kyle_7_18_17/"), os.path.join(os.environ["HOME"],pwd+"Kyle_3_7_17/")]
pd_list = []

for directory in directory_list:
    count=0
    for root,dirs,files in os.walk(directory):
        for file in files:
            if file.endswith(".csv"):
                pd_list.append(pd.read_csv(os.path.join(directory, file), delimiter=",", header=0))
                label = file.split("_S")
                for name in pd_list[count]["Name"]:
                    if name not in list_of_genes:
                        list_of_genes[name] = label[0]+ ":" + str(pd_list[count].loc[pd_list[count]["Name"] == name, "PolarityPerGene"].item()) +";"
                    else:
                        list_of_genes[name] = list_of_genes[name]+label[0] + ":" + str(pd_list[count].loc[pd_list[count]["Name"] == name, "PolarityPerGene"].item()) +";"
                count = count + 1
print(list_of_genes)
```

```python
#Convert to a dataframe
csv_columns = ['Name','PolarityPerGene']
combined_data = pd.DataFrame(list_of_genes.items(), columns=csv_columns)
combined_data
```

```python
csv_file = "combined.csv"
combined_data.to_csv(csv_file)
```

```python
pwd = "Zid-Lab/Analysis/Polarity/Kyle/"
directory = os.path.join(os.environ["HOME"],pwd+"Kyle_7_18_17/")
pd_list = []
label_list = []

def readInput(directory):
    for root,dirs,files in os.walk(directory):
        for file in files:
            if file.endswith(".csv"):
                temp = pd.read_csv(os.path.join(directory, file), delimiter=",", header=0, usecols = [1,2,3,4])
                pd_list.append(temp)
                label = file.split("_S")
                print("Read in file: " + label[0])
                label_list.append(label[0])


readInput(directory)
```
